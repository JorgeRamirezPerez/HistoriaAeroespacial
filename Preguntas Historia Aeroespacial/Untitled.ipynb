## 1) El padre de la Cosmonáutica, Konstantín Tsiolkovski, establece la ecuación del cohete. ¿En qué consiste?

La ecuación de Tsiolkoviski también es conocida como, ecuación del cohete clásico o ecuación del cohete ideales. Consiste en describir matemáticamente de los vehículos que pueden aplicar una aceleración a si mismo usando un empuje expulsando parte de su masa. 

Esto quiere decir que no se puede aplicar las ecuaciones clásicas de movimiento debido a que constantemente hay perdidas de masa, es decir, la masa es variable.

Las variables para tomar en consideración son: 

- Masa $m$ 
- Velocidad $v$ 
- Masa removida $m_{e}$
- Tiempo $\delta{t}$
- Incremento de velocidad $v$ a $v+dv$ puesto una cantidad variable de masa $\delta{m_{e}}$
  - *algo a tener en cuenta es que la velocidad de la masa removida* $v_{e}$ *como la veria un observador no varia debido a que al ser expulsada su masa continua constante. 
- los impuslos $\sum F_{cv}$ *representa la resultante de todas las fuerzas externas. (no incluye la fuerza que impulsa el volumen hacia adelante*

## 4) En Peenemünde se produce un gran avance para la electrónica aeroespacial. Para mejorar el éxito de las misiones que no llegaba al 30%, un matemático Erich Pieruschka plantea que los componentes mecánicos y electromecánicos del cohete siguen las mismas leyes que las sustancias radioactivas y los seres vivos. Y usando la tasa media de fallos a la hora fue capaz de justificar la probabilidad de éxito de las misiones, dando lugar al inicio del estudio de la fiabilidad. ¿En qué consistieron esos cálculos?


Para entender de lo que Hablaba el matématico Erich Pieruschka podemos empezar por definir los siguientes conceptos: 

- **Fallo:** Es toda alteración o interrupción en el cumplimiento de la función requerida.

- **Fiabilidad (de un elemento):** Es la probabilidad de que funcione sin fallos durante un tiempo (t) determinado, en unas condiciones ambientales dadas. 

- **Mantenibilidad:** Es la probabilidad de que, despues del fallo, sea reparado en un tiempo dado. 

- **Disponibilidad:** Es la probabilidad de que este en estado de funcionar (ni averiado ni en revisión) en un tiempo dado. 

lo que ha hecho Erich es implementar estos conceptos básicos que vienen de las leyes de "éxito" en los seres vivos como función natural para dar una estructuración a lo que lleva al comportamiento y como puede cambiar su forma de hacer las cosas para incrementar continuamente la fiabilidad y que disminuya el fallo. 

Formuló la fiabilidad de un sistema a partir de la fiabilidad de sus componentes, es decir, decia que era posible conocer la fiabilidad de un sistema completo a partir de conocer la fiabilidad de los componentes individuales que lo conforman. Esto se llevo posteriormente a en el sector militar de EEUU, para garantizar el funcionamiento de sistemas electrónicos y en la industria. 

la finalidad era, reducir riesgos, perdidas valiosas, verificar la calidad de un producto.

El Explorer 1 fue el primer Satélite en orbita de EEUU, que fue mandado en el cohete de 4 etapas **[juno 1 o Jupiter C](https://es.wikipedia.org/wiki/Juno_I)** este satelite fue diseñado y construido por el Jet propulsion Laboratory (JPL) la carga o Payload de este satelite era un intrumento de rayos cósmicos (lowa Cosmic Ray instrument) pero no poseia grabador de datos para poder comparar las señales que llegaban de el satelite. durante la operación, se descubrió que la tierra poseía un cinturo de radiación el cual lo conocemos como cinturon de Van Allen (quíen diseño junto al contador Geiger el detector de Iones de alta energía y electrónes, la carga de este satélite). 

Algunos de sus instrumentos eran: 

- un tubo de Geiger-Muller omnidireccional ANton 314 (detector de rayos cósmicos)

- Cinco sensores de tempreatura

- detector acustico para detectar impactos de micrometeoros

- antenas de alta y baja ganancia

- detector de malla para impactos

- baterias de mercurio que duraron 3 meses y representaban el 40% del peso de la carga útil.
